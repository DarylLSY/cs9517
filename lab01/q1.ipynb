```python
# 1. Contrast Stretching

import cv2
import numpy as np
import matplotlib.pyplot as plt

img_path = "pictures/Cat.png"

image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

def contrast(image):
    # For 8-bit RGB image
    a = 0
    b = 255
    c = 255
    d = 0

    (h, w) = image.shape

    # Calculate lowest and highest pixel brightness
    for i in image:
        for pixel in i:
            c = pixel if pixel < c else c
            d = pixel if pixel > d else d

    e = (b - a)/(d - c)

    # Iterate through image to adjust pixels
    for i in range(h):
        for j in range(w):
            image[i][j] = (image[i][j] - c) * e + a

    return image
```

```python
plt.imshow(contrast(image), 'gray')
```

```python
(h, w) = image.shape

min_coords = (0, 0)
max_coords = (0, 0)

Imin = 255
Imax = 0

for i in range(h):
    for j in range(w):
        intensity = image[i][j]
        if intensity < Imin:
            Imin = intensity
            min_coords = (i, j)
        if intensity > Imax:
            Imax = intensity
            max_coords = (i, j)

print('Minimum pixel value: ', Imin, ' at ', min_coords)
print('Maximum pixel value: ', Imax, ' at ', max_coords)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = 'images/Stars.png'

I = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
A = np.zeros(I.shape)
```

```python
def min_filtered(I, N):
    A = np.zeros(I.shape)
    h, w = I.shape

    m = int(N/2)
    for i in range(h - m):
        for j in range(w - m):
            A[i+m][j+m] = np.min(I[i:i+N, j:j+N])
    
    return A
```

```python
A = min_filtered(I, 3)
plt.imshow(A, cmap='gray', vmin=0, vmax=255)
```

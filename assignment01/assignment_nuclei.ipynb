```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = 'images/Nuclei.png'

I = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
N = 15
M = 1
```

```python
def filter(I, N, filter_type = 'min'):
    assert N % 2 != 0

    X = np.zeros(I.shape)
    h, w = I.shape

    m = int(N/2)
    for i in range(h - m):
        for j in range(w - m):
            if filter_type == 'min':
                X[i+m][j+m] = np.min(I[i:i+N, j:j+N])
            elif filter_type == 'max':
                X[i+m][j+m] = np.max(I[i:i+N, j:j+N])
    
    return X

def subtract(A, B, M):
    assert A.shape == B.shape

    X = np.zeros(A.shape)
    h, w = I.shape
    for i in range(h):
        for j in range(w):
            if M == 0:
                X[i][j] = A[i][j] - B[i][j]
            elif M == 1:
                X[i][j] = A[i][j] - B[i][j] + 255                

    return X
```

```python
if M == 0:
    A = filter(I, N, 'min')
    B = filter(A, N, 'max')
elif M == 1:
    A = filter(I, N, 'max')
    B = filter(A, N, 'min')

O = subtract(I, B, M)
```

```python
plt.imshow(I, cmap='gray', vmin=0, vmax=255)
```

```python
plt.imshow(B, cmap='gray', vmin=0, vmax=255)
```

```python
plt.imshow(O, cmap='gray', vmin=0, vmax=255)
```
